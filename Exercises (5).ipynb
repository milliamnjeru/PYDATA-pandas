### Step 5. Delete the column 'Unnamed: 0' and 'Id'

### Step 6. Is there more male or female names in the dataset?

```python
gender_counts = baby_names['Gender'].value_counts()
gender_counts
```

### Step 7. Group the dataset by name and assign to names

```python
names = baby_names.groupby('Name').sum()
names
```

### Step 8. How many different names exist in the dataset?

```python
num_names = baby_names['Name'].nunique()
print(f"Number of different names: {num_names}")
```

### Step 9. What is the name with most occurrences?

```python
most_occurrences = baby_names.groupby('Name')['Count'].sum().idxmax()
print(f"Name with the most occurrences: {most_occurrences}")
```

### Step 10. How many different names have the least occurrences?

```python
least_occurrences_count = baby_names.groupby('Name')['Count'].sum().min()
least_occurrences_names = baby_names.groupby('Name')['Count'].sum()
num_least_occurrences = least_occurrences_names[least_occurrences_names == least_occurrences_count].count()
print(f"Number of different names with the least occurrences: {num_least_occurrences}")
```

### Step 11. What is the median name occurrence?

```python
median_occurrence = baby_names.groupby('Name')['Count'].sum().median()
print(f"Median name occurrence: {median_occurrence}")
```

### Step 12. What is the standard deviation of names?

```python
std_deviation = baby_names.groupby('Name')['Count'].sum().std()
print(f"Standard deviation of name occurrences: {std_deviation}")
```

### Step 13. Get a summary with the mean, min, max, std and quartiles.

```python
summary = baby_names.groupby('Name')['Count'].sum().describe()
print(summary)
```
